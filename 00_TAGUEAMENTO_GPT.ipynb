```python
import logging
import time
import pandas as pd
from openai import OpenAI
from tqdm import tqdm

logger = logging.getLogger(__name__)

# Inicialize as listas para produtos e características
produtos_processados = []
caracteristicas = []

# Carregue o progresso existente, se houver
try:
    progresso_txt = pd.read_csv('progresso.txt')
    produtos_processados = progresso_txt['PRODUTOS'].tolist()
    caracteristicas = progresso_txt['CARACTERISTICAS'].tolist()
    print(f'progresso carregado com {len(produtos_processados)} produtos')
except FileNotFoundError:
    # Se o arquivo não existir, continue normalmente
    pass

# Inicialize o cliente OpenAI
client = OpenAI()

batch_size = 4
delay_seconds = 25  # Ajuste baseado no limite de taxa da API
products_to_process = df['id_produto'].dropna().unique()

# Exclua os produtos já processados da lista
products_to_process = [produto for produto in products_to_process if str(produto) not in str(produtos_processados)]

batches = [products_to_process[i:i + batch_size] for i in range(0, len(products_to_process), batch_size)]

start_time = time.time()

with tqdm(total=len(products_to_process)) as pbar:
    for batch in batches:
        for produto in batch:
            x = df.loc[df['id_produto'] == produto].reset_index(drop=True)
            grupo_produto = x['grupo_produto'].loc[0]
            url= [{
                    "type": "image_url",
                    "image_url": {"url": x['link'].loc[0]}
                    }]
            try:

                if grupo_produto == 'vestido':
                    grupo_message = tag.VESTIDO_CARAC
                elif grupo_produto == 'blusa':
                    grupo_message = tag.BLUSA_CARAC
                elif grupo_produto == 'calca':
                    grupo_message = tag.CALCA_CARAC
                elif grupo_produto == 'short':
                    grupo_message = tag.SHORT_CARAC
                elif grupo_produto == 'saia':
                    grupo_message = tag.SAIA_CARAC
                elif grupo_produto == 'blazer':
                    grupo_message = tag.BLAZER_CARAC
                else:
                    grupo_message = 'N/A'

                message = [{
                    "role": "user",
                    "content": [{
                        "type": "text",
                        "text": f" Você é um especialista de moda responsável pelo cadastro de atributos de roupas a partir de imagens."
                                f" Você está recebendo uma foto de um produto de roupa e sua tarefa é extrair atributos específicos da roupa e responder estritamente dessa forma: [{grupo_message}]."
                                f" Além da foto do produto, você receberá a informação de qual grupo esse produto pertence, podendo ser 'blazer', 'calca', 'blusa', 'saia', 'short' ou 'vestido'. Colete informações apenas da peça de vestuário referente a esse grupo."
                                " Observação 1: No grupo de produto 'blusa' você pode receber diferentes tipos de roupas superiores, como camisas, blusas, regatas, etc. Não se preocupe se a nomenclatura do grupo está correta, apenas colete os atributos da peça de roupa fornecida."                       
                                f" O grupo do produto que você recebeu é {grupo_produto}."
                                " Coloque a resposta final, conforme a formatação fornecida, entre chaves. Lembre-se de responder somente um valor por atributo. Escolha a melhor opção possível dentre as disponíveis para cada atributo."
                                " Você não dá avisos ou sugestões, apenas coleta os atributos e devolve estritamente conforme a formatação fornecida."
                    }] + url
                }]

                while True:
                    try:
                        # print(f'executando {produto}')
                        response = client.chat.completions.create(
                            model="gpt-4o-mini",
                            messages=message,
                            max_tokens=3000,
                        )
                        break
                    except Exception as error:
                        if "RateLimitError" in str(error) or "429" in str(error):
                            print(f'Rate limit error, waiting for {delay_seconds} seconds before retrying...')
                            time.sleep(delay_seconds)
                        else:
                            raise error
                pbar.update(1)
                produtos_processados.append(produto)
                caracteristicas.append(response.choices[0].message.content)

            except Exception as error:
                print(f'erro no produto:{produto} com {error}')
                logger.error('Failed to process product %s: %s', produto, repr(error))

        # Salva parcialmente o progresso em um txt
        progresso_txt = pd.DataFrame({'PRODUTOS': produtos_processados, 'CARACTERISTICAS': caracteristicas})
        progresso_txt.to_csv('progresso.txt', index=False)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Processamento concluído em {elapsed_time:.2f} segundos.")
```

```python
# Antes de colocar progresso.txt em um dataframe, tirar todas as quebras de linha que possam atrapalhar
produtos_categorizados = pd.read_csv('progresso.txt')
produtos_categorizados['CARACTERISTICAS'] = produtos_categorizados['CARACTERISTICAS'].str.replace('\n', '')
```

```python
def converte_dict(x):
        d = {}
        if not isinstance(x, list):
            return {x: None}
        for i in x:
            if not ":" in i:
                d[i] = None
            else:
                items = i.split(':')
                if len(items) == 2:
                    key, value = items
                    d[key.strip()] = value.strip()
                else:
                    key = items[0]
                    value = ':'.join(items[1:])
                    d[key.strip()] = value.strip()
        return d


# Função para dividir a coluna 'CARACTERISTICAS' em novas colunas
def split_caracteristicas(df, col):

    # Pegar apenas o que está dentro das chaves
    split_data = df[col].str.extract(r'{(.*)}', expand=False)

    # Dividir a coluna 'CARACTERISTICAS' por ';'
    split_data = split_data.str.split(';')
    
    # Converter cada item da lista em um dicionário
    split_data = split_data.apply(converte_dict)
    
    # Pegar todas as chaves
    keys = split_data.apply(lambda x: list(x.keys()))
    
    # Criar um conjunto de todas as chaves
    all_keys = set().union(*keys)

    for key in all_keys:
        # Criar uma nova coluna para cada chave
        df[key] = split_data.apply(lambda x: x.get(key))
    
    return df

produtos_categorizados = split_caracteristicas(produtos_categorizados, 'CARACTERISTICAS')
```
