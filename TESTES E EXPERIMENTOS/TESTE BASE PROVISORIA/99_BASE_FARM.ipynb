```python
import pandas as pd
import os
import csv
import requests

# Definindo o caminho do arquivo CSV e o diretório onde as imagens serão salvas
csv_file = './web-scrapping/farmrio_produtos.csv'
output_folder = 'base_imagens'

# Cria a pasta de saída, caso ela não exista
os.makedirs(output_folder, exist_ok=True)

# Função para baixar a imagem
def download_image(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as file:
            file.write(response.content)
        print(f'Imagem salva: {filename}')
    else:
        print(f'Erro ao baixar a imagem de {url}')

# Lê o arquivo CSV e processa cada linha
with open(csv_file, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        grupo = row['Grupo']
        titulo = row['Título']
        link = row['Imagem']
        
        # Modifica o link da imagem substituindo "-500-750" por ''
        link = link.replace('-500-750', '')

        # Cria o nome do arquivo seguindo o formato desejado
        filename = f"{grupo}_{titulo}.jpg".replace(' ', '_').replace('/', '-')
        filepath = os.path.join(output_folder, filename)

        # Verifica se a imagem já foi baixada
        if not os.path.exists(filepath):
            # Baixa a imagem se ainda não existir
            download_image(link, filepath)
        else:
            print(f'Imagem já existe: {filename}')
```

```python
# Verificando quantas imagens foram baixadas
print(f'{len(os.listdir(output_folder))} imagens baixadas')
```
