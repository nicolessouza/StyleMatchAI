Recebe classificacao_pesquisada, checks e retorna a lista de produtos na interseção.

```python
import pandas as pd
import os
```

```python
dict_busca = {
    "cor": "PRETO",
    "estampa": "liso",
    "filtrar_por_cor": True,
    "filtrar_por_estampa": True,
    "filtrar_por_vestido_comprimento_saia": False,
    "filtrar_por_vestido_decote_ou_gola": False,
    "filtrar_por_vestido_modelagem_saia": True,
    "filtrar_por_vestido_tipo_manga": True,
    "grupo_produto": "vestido",
    "vestido_comprimento_saia": "MIDI",
    "vestido_decote_ou_gola": "RETO",
    "vestido_modelagem_saia": "SOLTA",
    "vestido_tipo_manga": "ALCA"
}
```

```python
def search_products (dict_busca, path_datasets):
    # quais filtros foram selecionados? (filtar_por_{atributo} == True)
    filtros = [k for k, v in dict_busca.items() if 'filtrar_por_' in k and v == True]

    # replace filtrar_por por '' e transforma em lista
    filtros = [filtro.replace('filtrar_por_', '') for filtro in filtros]

    # Lendo df_classificacao
    df_classificacao = pd.read_parquet(f'{path_datasets}/df_classificacao.parquet')

    print(f"linhas: {df_classificacao.shape[0]}")

    # Filtrando df_classificacao para que coincida com todos os filtros selecionados
    for filtro in filtros:
        print("filtrando por", filtro, " == ", dict_busca[filtro])
        df_classificacao = df_classificacao[df_classificacao[filtro] == dict_busca[filtro]]
        print(f"linhas: {df_classificacao.shape[0]}")

    
    # Tirando colunas completamente nulas
    df_classificacao = df_classificacao.dropna(axis=1, how='all')

    return df_classificacao
```

```python
df_classificacao = search_products(dict_busca, './modules/datasets')
df_classificacao
```

```python
# transforma cada linha do output em uma saida
products = []
for index, row in df_classificacao.iterrows():
    product = {}
    for column in df_classificacao.columns:
        product[column] = row[column]
    products.append(product)

products
```
